```python
from mnist import Mnist
import cv2
from matplotlib import pyplot as plt
```

```python
path = 'Custom MNIST Samples/Digit 0/0_0.png'
model_path = 'mnist_model.h5'  # Specify your model path
mnist = Mnist(model_path)
true_digit = int(path.split('/')[2][0])
```

Read the image

```python
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
if img is None:
        raise ValueError(f"Image not found at path: {path}")
```

Resize and Display

```python
img_resized = cv2.resize(img, (28, 28))
plt.imshow(img_resized)
plt.show()
```

```python
predicted_digit = mnist.predict_digit(img_resized)

if predicted_digit == true_digit:
        print(f"Success: Image {path} for digit {true_digit} is recognized as {predicted_digit}.")
else:
        print(f"Fail: Image {path} for digit {true_digit} but the inference result is {predicted_digit}.")
```
