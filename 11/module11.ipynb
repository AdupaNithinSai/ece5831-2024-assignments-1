```python
# Importing necessary modules
from imdb import Imdb
from reuters import Reuters
from boston_housing import BostonHousing
```

```python
# Binary Classification - IMDB
print("Binary Classification using the IMDB dataset")
imdb_model = Imdb(num_words=10000)
imdb_model.prepare_data()
imdb_model.build_model()
imdb_model.train(epochs=20, batch_size=512)
imdb_model.model.save("saadhana-appikonda-imdb.h5")
imdb_model.plot_metrics()
imdb_evaluation = imdb_model.evaluate()
print(f"IMDB Evaluation: Loss = {imdb_evaluation[0]}, Accuracy = {imdb_evaluation[1]}")
```

```python
# Multiclass Classification - Reuters
print("Multiclass Classification using the Reuters dataset")
reuters_model = Reuters(num_words=10000)
reuters_model.prepare_data()
reuters_model.build_model()
reuters_model.train(epochs=20, batch_size=512)
imdb_model.model.save("saadhana-appikonda-reuters.h5")
reuters_model.plot_metrics()
reuters_evaluation = reuters_model.evaluate()
print(f"Reuters Evaluation: Loss = {reuters_evaluation[0]}, Accuracy = {reuters_evaluation[1]}")
```

```python
# Regression - Boston Housing
print("Regression using the Boston Housing dataset")
boston_housing_model = BostonHousing()
boston_housing_model.prepare_data()
boston_housing_model.build_model()
boston_housing_model.train(epochs=20, batch_size=32)
imdb_model.model.save("saadhana-appikonda-bh.h5")
boston_housing_model.plot_metrics()
boston_housing_evaluation = boston_housing_model.evaluate()
print(f"Boston Housing Evaluation: Loss = {boston_housing_evaluation[0]}, MAE = {boston_housing_evaluation[1]}")
```
