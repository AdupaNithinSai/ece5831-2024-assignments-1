# Dogs vs Cats Classification

```python
from dogs_cats import DogsCats

# Instantiate the class
dc = DogsCats()
print("Instance Created")
```

### Handling the Dataset
Creating the `dogs-vs-cats` directory using `make_dataset_folders()`
- 0 - 2,399 ==> validation
- 2,400 - 11,999 ==> Train
- 12,000 - 12,499 ==> Test

```python
# Create dataset folders
dc.make_dataset_folders('valid', 0, 2400)
dc.make_dataset_folders('train', 2400, 12000)
dc.make_dataset_folders('test', 12000, 12500)
print("Done")
```

### Creating a Dataset Object

```python
# Make datasets
dc.make_dataset()
print("Dataset Object Created")
```

### Building the Neural Network

```python
# Build the network
dc.build_network()
print("Network Built")
```

```python
# Display the model summary by calling it on a batch of data beforehand
for images, labels in dc.train_dataset.take(1):
    dc.model(images)
dc.model.summary()
```

### Training

```python
# Train the model
dc.train('model.saadhana-appikonda.dc-classifier.keras')
print("Model saved")
```

### Testing
Loading the test set, the model and testing the model over the test set

```python
def evaluate_test_set(model, test_dataset):
    test_loss, test_accuracy = model.evaluate(test_dataset)
    print(f'Test Loss: {test_loss}, Test Accuracy: {test_accuracy}')
    return 
```

```python
# Load and use the model for prediction
dc.load_model('model.saadhana-appikonda.dc-classifier.keras')
print("Model Loaded")
```

```python
# Calling the Evaluating function
evaluate_test_set(dc.model, dc.test_dataset)
```
